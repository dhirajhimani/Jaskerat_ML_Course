```python
number_of_tails = []
#prob = []

for vct in var_coin_tosses:
    
    count_tosses = np.random.binomial(n=1, p=.5, size=vct)
    number_of_tails.append(np.count_nonzero(count_tosses))
    #prob.append(number_of_tails/vct)
```

```python
number_of_tails
```

```python
#prob
```

```python
nt = np.array(number_of_tails)
```

```python
import matplotlib.pyplot as plt
```

```python
relative_frequency = nt/var_coin_tosses
```

```python
relative_frequency
```

```python
plt.stem(var_coin_tosses, relative_frequency)
```
