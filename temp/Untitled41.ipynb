```python
labels = data['diagnosis']
```

```python
data.drop([data.columns[0],data.columns[1],data.columns[32]],axis=1,inplace=True)
```

```python
data_dash = data - data.mean()
```

```python
data_dash
```

```python
sigma_dash = np.array(data_dash.cov())
```

```python
sigma_dash.shape
```

```python
factorized_matrices = np.linalg.svd(sigma_dash)
```

```python
data = np.array(data)
```

```python
def PCAed_data(data,variance_to_preserve):
    
    L = 1
    
    eig_values_total = np.sum(factorized_matrices[1])
    
    eig_values_sum = 0
    
    for eig_val in factorized_matrices[1]:
        
        if eig_values_sum/eig_values_total >= variance_to_preserve:
            
            break
            
            
        eig_values_sum = eig_values_sum + eig_val
            
        L = L + 1
            
    M = factorized_matrices[0][:,0:L]
    
    print(M.shape)
    
    return np.matmul(data,M)
```

```python
reduced_data = PCAed_data(data,0.999999)
```

```python
reduced_data_cov = np.cov(reduced_data,rowvar=0)
```

```python
reduced_data_cov.shape
```

```python
import scipy.stats as s 
```

```python
def calculate_posterior_probability(data):
    
    numerator = s.multivariate_normal.pdf(data,M_data_mean_vector,M_data_cov)
    
    Likelihood_B = s.multivariate_normal.pdf(data,B_data_mean_vector,B_data_cov)
    
    denominator = numerator + Likelihood_B
    
    posterior_M = numerator/denominator
    
    return posterior_M > 0.5
```

```python
predicted_outcome = calculate_posterior_probability(np.array(reduced_data.iloc[:,0:9]))
```

```python
predicted_outcome = np.uint8(predicted_outcome)
```
