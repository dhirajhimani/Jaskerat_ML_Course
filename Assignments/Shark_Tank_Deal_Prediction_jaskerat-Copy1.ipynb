```python
def set_of_words(desc):
    
    desc_list = desc.split(sep=" ")
    
    desc_set = set(desc_list)
    
    return desc_set
```

```python
pos_class_data = new_raw_data[new_raw_data['Deal_cracked_or_not'] == 1]
```

```python
neg_class_data = new_raw_data[new_raw_data['Deal_cracked_or_not'] == 0]
```

```python
pos_class_training_data = pos_class_data.iloc[0:int(0.75*len(new_raw_data))//2]
```

```python
def frequency_determination(desc):
    
    one_hot_encoded_list = []

    for unique_words in list(unique_bag_of_words):
        
        one_hot_encoded_list.append(unique_words in desc)

    return one_hot_encoded_list
```

```python
one_hot_encoded_vectors_pos = []

one_hot_encoded_vectors_pos.extend(map(lambda x: frequency_determination(x),pos_class_training_data.iloc[:,0]))
```

```python
one_hot_encoded_vectors_pos = np.array(one_hot_encoded_vectors_pos)
```

```python
cracked_deal_word_frequencies = np.sum(one_hot_encoded_vectors_pos,axis=0)
```

```python
not_cracked_deal_word_relative_frequencies += (1/(neg_class_training_data.shape[0]+len(unique_bag_of_words)))*(not_cracked_deal_word_relative_frequencies == 0)
```

```python
D_not_cracked_deal = dict(zip(list(unique_bag_of_words),list(not_cracked_deal_word_relative_frequencies)))
```

```python
D_not_cracked_deal
```

```python
def calculate_likeli_probability(word_relative_frequencies,data):

    likeli_prob = []

    for word in list(data):
    
        likeli_prob.append(word_relative_frequencies[word])
        
    return np.prod(np.array(likeli_prob))
```

```python
likeli_probabilities_cracked = []

likeli_probabilities_cracked.extend(map(lambda x,y: calculate_likeli_probability(x,y),[D_cracked_deal]*len(testing_data),testing_data.iloc[:,0]))
```

```python
likeli_probabilities_cracked = np.array(likeli_probabilities_cracked)
```

```python
likeli_probabilities_not_cracked = []

likeli_probabilities_not_cracked.extend(map(lambda x,y: calculate_likeli_probability(x,y),[D_not_cracked_deal]*len(testing_data),testing_data.iloc[:,0]))
```

```python
likeli_probabilities_not_cracked = np.array(likeli_probabilities_not_cracked)
```

```python
posterior_cracked = likeli_probabilities_cracked/(likeli_probabilities_cracked + likeli_probabilities_not_cracked)
```

```python
posterior_cracked
```

```python
predicted_classes = np.uint8(posterior_cracked > 0.5)
```

```python
predicted_classes
```

```python
actual_classes = np.array(testing_data.iloc[:,1])
```

```python
correct_count_test = np.count_nonzero(np.equal(predicted_classes,actual_classes))
```
