The first task is to create a new folder named 'Cross Validation' inside the Parent folder named 'DevnagriHandwrittenCharacterDataset' along with 'Train' and 'Test':

```python
os.makedirs("./DevanagariHandwrittenCharacterDataset/Cross Validation")
```

Inside './DevnagriHandwrittenCharacterDataset/Cross Validation' folder, you have to create all the 46 folders with the exact names as the 46 folders present inside './DevnagriHandwrittenCharacterDataset/Train': 

```python
os.listdir("./DevanagariHandwrittenCharacterDataset/Train")
```

After creating 46 folders inside './DevnagriHandwrittenCharacterDataset/Cross Validation' with the same names as mentioned above, you have to move (cut) 25% of the images available in each of the folders in './DevnagriHandwrittenCharacterDataset/Train' to the corresponding folder with the same name in './DevnagriHandwrittenCharacterDataset/Cross Validation'. As an example, let's pick one folder named 'character_10_yna' in './DevnagriHandwrittenCharacterDataset/Train' that is:

'./DevnagriHandwrittenCharacterDataset/Train/character_10_yna'. 

We know that there are 1700 images in this folder and other such folders of different Devnagri Characters. Now, we have to move (cut) not copy 25% random taken images from this folder, that is 0.25X1700 = 425 images to the following folder:

'./DevnagriHandwrittenCharacterDataset/Cross Validation/character_10_yna'. 

We have to do the same procedure for all the 46 folders in the two parent folders: 'Train' and 'Cross Validation'. 

Therefore, after moving 25% of the images from each of the folders in 'Train' folder to each of the folders in 'Cross Validation' folder, the 'Train' folder is left with each folder inside of it having only 1700 - (0.25X1700) = 1275 images in each folder inside 'Train' folder. 

As there are 46 folders in 'Train' folder, the total number of images in 'Train' folder will be: 46X1275 = 58,650 images. 

Let's try to have a look on one of these images in one of the folders by importing matplotlib.pyplot library and using plt.imread() function to read the image. 

```python
import matplotlib.pyplot as plt
```

```python
random_image = plt.imread(os.path.join("./DevanagariHandwrittenCharacterDataset","Train/character_10_yna/10542.png"))

plt.imshow(random_image,cmap='gray')
```

When an image is read in Python using plt.imread() function, it is read as a numpy array or in layman terms, you can say a matrix. Let's see how many rows and columns of pixels in any random image:

```python
print(random_image.shape)
```

So, there are 32X32 = 1024 pixels in an image and an image after reading can be reshaped into (1,1024) shape. 

This can be done as:

```python
random_image = random_image.reshape(1,random_image.shape[0]*random_image.shape[1])
```

```python
random_image.shape
```

All of 58,650 images can be reshaped in such a manner and if they are stacked vertically one after the other rowwise, then we get 58,650 rows where each row corresponds to the each image and all the columns in each row corresponds to all the pixels in an image and as there are 58,650 rows so after vertically stacking each row, we get a numpy array of shape (58650,1024).
