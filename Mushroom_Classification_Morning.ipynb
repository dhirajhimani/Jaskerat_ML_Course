```python
def calculate_p_cap_shape (dataframe):
    
    p_cap_shape_on = dict(dataframe['cap-shape'].value_counts())
    
    print(p_cap_shape_on)

    value_counts = np.array(list(p_cap_shape_on.values()))
    
    print(value_counts)

    rf = value_counts/dataframe.shape[0]
    
    print(rf)

    p_cap_shape_on = dict(zip(p_cap_shape_on.keys(),rf))
    
    return p_cap_shape_on
```

```python
e_training_data.shape
```

```python
p_cap_shape_on_e = calculate_p_cap_shape(e_training_data)
```

```python
p_cap_shape_on_p = calculate_p_cap_shape(p_training_data)
```

```python
p_cap_shape_on_e['c'] = 1/(e_training_data.shape[0] + training_data['cap-shape'].unique().shape[0])
```

```python
def predicted_category(x):
    
    likelihood_e = p_cap_shape_on_e[x]
    
    likelihood_p = p_cap_shape_on_p[x]
    
    numerator = likelihood_p * p_cap[1]
    
    denominator = numerator + (likelihood_e * p_cap[0])
    
    posterior_p = numerator/denominator
    
    return int(posterior_p > 0.5)
```

```python
predicted_answers = []

for cap_shape in cross_validation['cap-shape']:
    
    predicted_answers.append(predicted_category(cap_shape))
```
